```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/processed/hotel_bookings_final.csv")

df["reservation_status_date"] = pd.to_datetime(
    df["reservation_status_date"]
)
```

## Creating a SQL Server Connection

```python
from sqlalchemy import create_engine

server = r"Karthikk\SQLEXPRESS"
database = "TourismAnalyticsDB"

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+18+for+SQL+Server"
    "&trusted_connection=yes"
    "&TrustServerCertificate=yes"
)

engine = create_engine(connection_string)
```

```python
with engine.connect() as conn:
    print("Connected to SQL Server successfully!")
```

```python
df.to_sql(
    "hotel_bookings_final",
    engine,
    if_exists="replace",
    index=False
)
```

## Uploading the Data Frame

```python
df.to_sql(
    name="hotel_bookings_final",
    con=engine,
    if_exists="replace",
    index=False
)

print("Data uploaded successfully!")
```
