## Data Cleaning

### Filling Missing Values

```python
# Fill missing company values
df["company"] = df["company"].fillna("Not Applicable")

# Fill missing agent values
df["agent"] = df["agent"].fillna("Not Applicable")

# Fill missing country values
df["country"] = df["country"].fillna("Unknown")

# Fill missing children values
df["children"] = df["children"].fillna(0)
```

### Verifying Missing Values

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

### Converting Date from object type to datetime64

```python
df["reservation_status_date"] = pd.to_datetime(
    df["reservation_status_date"]
)
```

### Converting Childern datatypr from float64 to int64

```python
df["children"] = df["children"].astype(int)
```
